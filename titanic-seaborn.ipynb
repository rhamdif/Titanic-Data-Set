###### 1- Display All Survived And Died chart

```python
data = myData.copy()
data["survived"]= data["survived"].apply(lambda x : "Survived" if x == 1 else "Died")
```

```python
sns.countplot(x=data["survived"])
plt.title("Display Survived and Died")
plt.xlabel("Classes")
plt.ylabel("count")
plt.show()
```

```python
sns.countplot(data=data, hue="survived",x="sex")
plt.title("Display Survived and Died Grouped By Gender")
plt.xlabel("Classes")
plt.ylabel("count")
plt.show()
```

### 4- Display Survived And Died in Each PClass chart

```python
def conditions(x):
    if x == 1.0  :
        return "1st Class"
    elif x == 2.0  :
        return "2nd Class"
    else:
        return "3rd Class"
data["pclass"] = data["pclass"].apply(lambda x : conditions(x) )
sns.countplot(data=data, hue="survived",x="pclass")
plt.title("Display Survived and Died Grouped By PClass")
plt.xlabel("Classes")
plt.ylabel("count")
plt.show()
```

### 5- Display Survived And Died in age as ["Adult","Child","Old"] chart

```python
def conditions(x):
    if x < 12 :
        return "child"
    elif x >= 12 and x < 45:
        return "Adult"
    else:
        return "old"
         
data["age"] = data["age"].apply(lambda x : conditions(x) )
sns.countplot(data=data, hue="survived",x="age")
plt.title("Survived and Died Grouped By age[child, adult, old]")
plt.xlabel("Classes")
plt.ylabel("count")
plt.show()
```

### 6- Display Survived And Died in Male->age as ["Adult","Child","Old"] chart


```python
g1 = data[["age","sex","survived"]].groupby(data["sex"][data["sex"]=="male"]).get_group("male")  
sns.countplot(data=g1, hue="survived",x="age")
plt.title("Survived and Died Grouped By gender[male] and age[child, adult, old]")
plt.xlabel("Classes as per male")
plt.ylabel("count")
plt.show()
```

### 7- Display Survived And Died in Female->age as ["Adult","Child","Old"] chart


```python
g2 = data[["age","sex","survived"]].groupby(data["sex"][data["sex"]=="female"]).get_group("female")  
sns.countplot(data=g2, hue="survived",x="age")
plt.title("Survived and Died Grouped By gender[female] and age[child, adult, old]")
plt.xlabel("Classes as per female")
plt.ylabel("count")
plt.show()
```
